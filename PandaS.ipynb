```python
#crear una seria
serie=pd.Series([2,3,5,6,4],index=['a','b','c','d','e'])
serie
```

```python
#Borra seie o fila
```

```python
serie=pd.Series.drop(['a','b'])
serie
```

```python
#Seleccionar filas o series
serie1=serie[['b','d']]
serie1
```

```python
#Asignacion
serie['b']=100
serie
```

```python
#operadore
serie4=(serie+serie1)
serie4
```

```python
#multiplicacion
serie_mul=serie.mul(serie4,serie)
serie_mul
```

```python
#division

serie_div=serie.div(serie4,serie)
serie_div
```

```python
#resta
serie_res=serie.subtract(serie4,serie1)
serie_res
```
